```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
import pandas as pd
from time import sleep

options = webdriver.ChromeOptions()


```

```python
options.add_argument("--start-maximized")

driver = webdriver.Chrome(executable_path='./chromedriver',options=options)
driver.get("https://voterlist.election.gov.np/bbvrs1/index_2.php")

sleep(2)

def formFill(pradesh,jilla,napa,wada,kendra):

    province = Select(driver.find_element_by_id('state'))

    province.select_by_value(str(pradesh))

    sleep(2)

    district = Select(driver.find_element_by_id('district'))

    district.select_by_value(str(jilla))

    sleep(1)

    vdc_mun = Select(driver.find_element_by_id('vdc_mun'))

    vdc_mun.select_by_value(str(napa))

    sleep(2)

    ward = Select(driver.find_element_by_id('ward'))

    ward.select_by_value(str(wada))

    sleep(1)

    center = Select(driver.find_element_by_id('reg_centre'))

    center.select_by_value(str(kendra))

    sleep(2)

    driver.find_element_by_id('btnSubmit').click()



formFill(5,77,5474,1,5960)

sleep(6)

entries = Select(driver.find_element_by_name('tbl_data_length'))
entries.select_by_value('100')

sleep(5)


df = pd.DataFrame()

#since there are 42xx rows so it needs 43 pages..

#the first 42 loops will add data and also click next page button
#43rd page only appends data but does not click next page button 
for i in range(1,44):
    sleep(2)
    html = driver.find_element_by_id('tbl_data_wrapper')
    htl = html.get_attribute('innerHTML')
    table = pd.read_html(htl)
    new = pd.DataFrame(table[0])
    df = df.append(new)
    sleep(1)
    if i < 43:
        driver.find_element_by_id('tbl_data_next').click()

df


```
